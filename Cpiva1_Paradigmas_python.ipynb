### **Analise de palavras dos poemas de mario quitana no intuito de montar o dicionário de aprendizagem trilha 1.**

Efetuei analise das palavras utilizadas pelo escritor que tanto tenho gosto por leitura Mario Quitana, achei interessante fazer analise para saber as palavras que o autor mais utilizava em seus poemas que falam sobre sensações e sentimentos.
Link utilizado foi adquirido no site:https://poesiaspoemaseversos.com.br/mario-quintana-poemas/


```python
import urllib.request

texto = []

f = open ("mario_quitana.txt",'r') # para arquivos locais
for line in f:
  texto.append(line)


  for i in range(len(texto)):
    texto[i] = texto[i].lower() 
    texto[i] = texto[i].replace('\n','')
    texto[i] = texto[i].replace('.','')
    texto[i] = texto[i].replace(',','')
    texto[i] = texto[i].replace('(','')
    texto[i] = texto[i].replace(')','')
    texto[i] = texto[i].replace('?','')
    texto[i] = texto[i].replace('\'','') 
    texto[i] = texto[i].replace('que','')
    texto[i] = texto[i].replace('um','')    
```

# **Criando dicionário**



```python
mydict = {}                             

for line in texto:

  line = line.lower()                   
  words = line.split()                 
  #print(words)

  for word in words:                     
    if word not in mydict.keys():        
      mydict[word] = 1                    
    else:                                 # se a entrada já existe 
      mydict[word] =  mydict[word] + 1    

print(mydict) 
```

```python

import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

df = pd.DataFrame(mydict.items(), columns=['word', 'count']).sort_values('count',ascending=False)
df = df[df['count'] > 4] 

plt.figure(figsize=(24,10))
mpl.style.use(['seaborn'])
sns.barplot(x=df.word,y=df['count'])
plt.xticks(rotation=90) 

plt.show()
```
